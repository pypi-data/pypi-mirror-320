```python
# --------------------------------------------------------------------------------
# Задача 2
# --------------------------------------------------------------------------------
# 1) Вычисление выборочного коэффициента корреляции Пирсона между X и Y
# --------------------------------------------------------------------------------

import numpy as np

corr_pearson = np.corrcoef(X, Y)[0, 1]
print("1) Выборочный коэффициент корреляции Пирсона:", corr_pearson)
```

```python
# --------------------------------------------------------------------------------
# Задача 2 (продолжение)
# --------------------------------------------------------------------------------
# 2.1) Проверка гипотезы о равенстве средних значений (двусторонний тест, без предположения о равенстве дисперсий)
#     Вычисляем p-value.
# 2.2) На уровне значимости 0.01, проверяем, различаются ли средние значения показателей у фирм.
#     Если p-value < 0.01 => 1, иначе => 0
# --------------------------------------------------------------------------------

from scipy.stats import ttest_ind

ttest_res = ttest_ind(X, Y, equal_var=False)  # Welch's t-test
p_value_means = ttest_res.pvalue

print("2.1) P-value (проверка равенства средних, двусторонняя):", p_value_means)

alpha_means = 0.01
diff_means_result = 1 if p_value_means < alpha_means else 0
print("2.2) На уровне значимости 0.01, различаются ли средние? (1 - да, 0 - нет):", diff_means_result)
```

```python
# --------------------------------------------------------------------------------
# Задача 3
# --------------------------------------------------------------------------------
# 3.1) Проверка гипотезы о равенстве дисперсий показателей двух фирм
#      Альтернативная гипотеза: дисперсия показателя больше у первой фирмы (X)
#      Используем F-тест: F = Sx^2 / Sy^2
#      p-value = 1 - cdf_F(F, df1, df2), так как проверяем одностороннюю гипотезу var(X) > var(Y)
#
# 3.2) На уровне значимости 0.1, проверяем, является ли дисперсия у X строго больше?
#      Если p-value < 0.1 => 1, иначе => 0
# --------------------------------------------------------------------------------

from scipy.stats import f

# Оценки выборочных дисперсий (ddof=1 для несмещенной оценки)
Sx2 = np.var(X, ddof=1)
Sy2 = np.var(Y, ddof=1)
F_stat = Sx2 / Sy2

dfn = len(X) - 1  # число степеней свободы для X
dfd = len(Y) - 1  # число степеней свободы для Y

# Для альтернативы var(X) > var(Y):
p_value_var = 1 - f.cdf(F_stat, dfn, dfd)

print("3.1) P-value (проверка о том, что var(X) > var(Y)):", p_value_var)

alpha_var = 0.1
diff_var_result = 1 if p_value_var < alpha_var else 0
print("3.2) На уровне значимости 0.1, var(X) > var(Y)? (1 - да, 0 - нет):", diff_var_result)
```
