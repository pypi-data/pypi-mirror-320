```python
from src.mcmctoolbox.mcmc import HamiltonianMonteCarlo, MCMCFactory
from scipy.stats import norm
import numpy as np
```

```python
log_target_pdf = lambda x: norm.logpdf(x, loc=0, scale=1)
grad_log_target_pdf = lambda x: -x
initial_sample = np.array([0.0])
```

```python
hmc = MCMCFactory.create(
    "hmc",
    log_target_pdf=log_target_pdf,
    grad_log_target_pdf=grad_log_target_pdf,
    init_sample=initial_sample,
)
```

```python
hmc.sample()
```

```python
hmc.plot()
```
